```python
import pandas as pd
import numpy as np
import gensim
model = gensim.models.Word2Vec.load("data//models//w2v_model_w9m30.model")
```

```python
with open('data//new_age_topic.txt' , encoding = 'ANSI') as f:
    lines = f.readlines()
lines
new_age_topics = {}
for line in lines:
    list_line = line.replace("\n", "").split(' ')
    new_age_topics[list_line[0]] = list_line[1:]
```

```python
with open('data//self_enhancement_topic.txt' , encoding = 'ANSI') as f:
    lines = f.readlines()
lines
self_enhancement_topics = {}
for line in lines:
    list_line = line.replace("\n", "").split(' ')
    self_enhancement_topics[list_line[0]] = list_line[1:]
```

```python
def average_vector(model, words_list):
    n = len(words_list)
    sum_vector = model.wv[words_list[0]]
    for word in words_list[1:]:
        word_vector = model.wv[word]
        sum_vector = sum_vector +  word_vector
    return sum_vector/n
```

```python
ix = list(new_age_topics.keys())
columns = list(self_enhancement_topics.keys())
data = pd.DataFrame(columns = columns, index = ix)
```

```python
for key1 in columns:
    for key2 in ix:
        
        sentence_1 = new_age_topics[key2]
        sentence_2 = self_enhancement_topics[key1]
        vector_1 = average_vector(model,sentence_1)
        vector_2 = average_vector(model,sentence_2)
        data[key1][key2] = np.dot(vector_1, vector_2) /( np.linalg.norm(vector_1)*np.linalg.norm(vector_2))
        
data.astype('float').style.background_gradient(cmap='Blues')
```
